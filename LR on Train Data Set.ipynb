```python
# Assuming 'y' is the target variable and 'x' is the feature
X = df[['x']]  # Selects the 'x' column as features
y = df['y']    # Selects the 'y' column as the target
```

```python
# Import train_test_split function from sklearn.model_selection
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
# Import LinearRegression function from sklearn.model_selection
from sklearn.linear_model import LinearRegression
```

```python
# LinearRegression fits a linear model with coefficients w = (w1, , wp) to minimize the residual sum of squares between the observed targets
# in the dataset, and the targets predicted by the linear approximation.
lm = LinearRegression()
```

```python
# Train the Linear Regression Classifer
lm.fit(X_train,y_train)
```

```python
# Predict values based on linear model object.
predictions = lm.predict(X_test)
y_test_clean = np.nan_to_num(y_test)
predictions_clean = np.nan_to_num(predictions) #replace NaN values in both y_test and predictions with zero
```

```python
plt.scatter(y_test, predictions)
#add label to scatter plot
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

```python
# Importing metrics from sklearn
from sklearn import metrics
```

```python
# Show the values of MAE, MSE, RMSE
print('MAE:', metrics.mean_absolute_error(y_test_clean, predictions_clean))
print('MSE:', metrics.mean_squared_error(y_test_clean, predictions_clean))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test_clean, predictions_clean)))
```
